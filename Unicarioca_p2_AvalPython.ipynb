```python
# 3. Importando o repositórios dos dados em um dataframe (data)
data = pd.read_table("http://archive.ics.uci.edu/ml/machine-learning-databases/image/segmentation.data", sep=",", header=[2], skip_blank_lines=True)
display(data)
```

```python
# 8. Identificando as colunas (características) dados NA
data.isnull().sum()
```

```python
# 8.1. Pela resposta acima, não há necessidade de tratamento da dados NA
#      Mas caso houvesse (e para cumprir o enunciado da questão)...
#      Limpando a coluna "SATURATION-MEAN"
saturation = data[["SATURATION-MEAN"]].dropna()
display(saturation)
```

```python
sns.displot(saturation, binwidth = 0.05, kde=True, rug = True, legend = False)
plt.title("Distribuição da Saturação (Já Normalizada)")
#plt.legend()
plt.style.use('ggplot')
```

#### Conclusão
###### A variável testada "SATURATION-MEAN" apresenta distribuição assimétrica positiva, sendo 
###### 0.25<Moda<0.30 < Mediana ~0.36 < Média ~0.42,
###### e leptocúrtica (alta concentração em torno da moda), com coeficiente de curtose ~0.18.
###### Possível análise preditiva do comportamento desta variável poderia partir de modelagem baseada em Distribuição de Poisson, a partir dos parâmetros levantados na análise exploratória. Porém, na hora da modelagem, deve-se atentar para aumento da concentração na cauda da distribuição (próximo a 1.0).
